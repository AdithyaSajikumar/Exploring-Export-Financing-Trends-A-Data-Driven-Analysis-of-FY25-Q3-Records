***Project Overview***

**Domain**: Banking and Finance

**Objective**: This project analyzes the FY25 Q3 export financing dataset from the U.S. Export-Import Bank (Data.gov).The goal is to clean, explore, and visualize the data to uncover key insights such as approval trends, top countries and products financed, program comparisons, support for diverse businesses, seasonal patterns, and financial risk exposure.By turning raw records into structured analysis, the project highlights how finance and banking analytics can guide decision-making in international trade.

**Problem Definition**: International trade financing involves complex approval processes, diverse business categories, and varying financial risks.  
Decision-makers often struggle to identify clear patterns in approvals, country-level financing, product trends, and support for small or minority-owned businesses. This project addresses the problem of turning raw government financing records into actionable insights. By applying data cleaning, exploratory analysis, and visualization techniques, the project aims to:
- Reveal approval trends and seasonal patterns  
- Compare financing across countries, products, and programs  
- Assess support for diverse businesses (small, women-owned, minority-owned)  
- Highlight financial risk exposure through outstanding loan analysis  
The outcome provides a structured view of export financing trends, enabling better understanding of how government support impacts international trade and business growth.
